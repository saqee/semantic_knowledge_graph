```python
faulty_data=data.faulty
faulty_data
```

```python
without_faulty_data=data.drop(['faulty','location','equipment'],axis=1)
without_faulty_data
```

```python
from sklearn.feature_selection import SelectKBest
```

```python
from sklearn.feature_selection import f_classif
```

```python
FEATURES=SelectKBest(score_func=f_classif)
```

```python
FEATURES.fit(without_faulty_data,faulty_data)
```

```python
FEATURES.scores_
```

```python
data_col=pd.DataFrame(data=FEATURES.scores_,columns=['score'])
data_col
```

```python
name_col=pd.DataFrame(data=without_faulty_data.columns,columns=['Features'])
name_col
```

```python
merge_data=pd.concat([name_col,data_col],axis=1)
```

```python
merge_data
```

```python
merge_data.nlargest(3,'score')
```
